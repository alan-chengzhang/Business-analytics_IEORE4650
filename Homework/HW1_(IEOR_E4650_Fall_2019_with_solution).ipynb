```
#This method will report one of the highest CP pokemon within each type group
#Give full points as long as students can find the correct highest total for each pokemon type


#Version 1:
Pokemon_type1=Pokemon[["Name","Type_1","Total"]].rename(columns={"Type_1":"Type"})
Pokemon_type2=Pokemon[["Name","Type_2","Total"]].rename(columns={"Type_2":"Type"})
Pokemon_combined=pd.concat([Pokemon_type1,Pokemon_type2]).dropna()
Pokemon_grouped=Pokemon_combined.sort_values(by=["Total"],ascending=False).groupby("Type").first()
Pokemon_grouped
```

```
#Version 2:
Pokemon_type1=Pokemon[["Name","Type_1","Total"]].rename(columns={"Type_1":"Type"})
Pokemon_type2=Pokemon[["Name","Type_2","Total"]].rename(columns={"Type_2":"Type"})
Pokemon_combined=pd.concat([Pokemon_type1,Pokemon_type2]).dropna()
Pokemon_combined.groupby("Type").max()
```

```
##An even better answer that will take care of ties

Pokemon_type1=Pokemon[["Name","Type_1","Total"]].rename(columns={"Type_1":"Type"})
Pokemon_type2=Pokemon[["Name","Type_2","Total"]].rename(columns={"Type_2":"Type"})
Pokemon_combined=pd.concat([Pokemon_type1,Pokemon_type2]).dropna()
Type_Total=Pokemon_combined.groupby("Type").max()
Type_Total_cleaned=pd.DataFrame({"Type": Type_Total.index.values, "max":Type_Total["Total"].values})

Pokemon_combined.merge(Type_Total_cleaned, left_on=["Type","Total"],right_on=["Type","max"]).drop(columns="max")
```
